```python
#Project
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC, LinearSVC
from sklearn import svm

from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics  import confusion_matrix, classification_report


twitter_file = "final_final_final_annotated_version.csv"

# Read in the csv into a pandas data frame.
dataset = pd.read_csv(twitter_file,encoding='utf-8')

print(dataset.shape)

#drop the NaN with 
#twitter_data.dropna(inplace=True)

print ("Number of observations :", len(dataset.index))
print ("Number of columns : ", len(dataset.columns))
print ("Headers : ", dataset.columns.values)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec = TfidfVectorizer()
X = tf_vec.fit_transform(data)
X.shape
```

```python
target = dataset["sentiment"]
```

## 3 Classifiers: Data [Tweets] vs Target [Sentiment]

```python
# Training Phase
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.20, random_state=10)
```

```python
#Classifier 1: BernoulliNB
from sklearn.naive_bayes import BernoulliNB

classifier_Ber = BernoulliNB().fit(X_train,y_train)
print("\nBernoulliNB accuracy: %.2f" %classifier_Ber.score(X_test, y_test))

results_Ber = cross_val_score(classifier_Ber, X_train,y_train, cv=10)
print("\n10-fold cross-validation:")
print(results_Ber)
print("The average accuracy of the BernoulliNB classifier is : %.2f" % np.mean(results_Ber))
print("\nConfusion matrix of the BernoulliNB classifier:")
predicted_Ber = classifier_Ber.predict(X_test)
print(confusion_matrix(y_test,predicted_Ber))
print("\nClassification_report of the BernoulliNB classifier:")
print(classification_report(y_test,predicted_Ber))
print("----------------------------------------------------------------------------")
```

```python
#Classifier 2: SVM

classifier_svm = svm.SVC(kernel='linear', C=1).fit(X_train,y_train)
print("SVM accuracy: %.2f"%classifier_svm.score(X_test, y_test))

results_svm = cross_val_score(classifier_svm, X,target, cv=10)
print("\n10-fold cross-validation:")
print(results_svm)
print("The average accuracy of the SVM classifier is : %.2f" % np.mean(results_svm))
print("\nConfusion matrix of the SVM classifier:")
predicted_svm = classifier_svm.predict(X_test)
print(confusion_matrix(y_test,predicted_svm))
print("\nClassification_report of SVM classifier:")
print(classification_report(y_test,predicted_svm))
print("----------------------------------------------------------------------------")
```

```python
#Classifier 3: Logistic 
classifier_log=  LogisticRegression().fit(X_train,y_train)
print("Logisitic Accuracy: %.2F"%classifier_log.score(X_test, y_test))

results_log = cross_val_score(classifier_log, X,target, cv=10)
print("\n10-fold cross-validation:")
print(results_log)
print("The average accuracy of the SVM classifier is : %.2f" % np.mean(results_log))
print("\nConfusion matrix of the SVM classifier:")
predicted_log= classifier_log.predict(X_test)
print(confusion_matrix(y_test,predicted_log))
print("\nClassification_report of SVM classifier:")
print(classification_report(y_test,predicted_log))
print("----------------------------------------------------------------------------")
```

```python
f = open('tweets.txt', encoding='utf-8')
```

```python
stopwords_list = stopwords.words('arabic')

vectorizer = CountVectorizer(stop_words = stopwords_list)
transformer = TfidfTransformer()
```

```python
trainVectorizerArray = vectorizer.fit_transform(f).toarray()
```

```python
tf_feature_names = vectorizer.get_feature_names()
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
n_samples = 2000
n_features = 1000
n_components = 20
n_top_words = 7
lda = LatentDirichletAllocation(n_components=n_components, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
lda.fit(trainVectorizerArray)
```

```python
print_top_words(lda,tf_feature_names,n_top_words)
```
